```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer(strategy='constant') 

# https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html
# Fills with fill_value (default = 0 for numerical, 'missing_value' for object)
# Other options are mean, median, most_frequent
```

```python
data = pd.DataFrame(
                    [
                        [20, 5, 100], [21, 9, 300], [22, 11, 400]#, [23, 12, np.NaN]]
                        , [np.NaN, np.NaN, np.NaN]]
                        , columns=['temp', 'mist', 'rain']
                        , index=['jan', 'feb', 'mar', 'apr']
                   )
data
#data.dropna(axis = 1, subset = ['apr'])
```

```python
cols_with_missing = [col for col in data.columns
                     if data[col].isnull().any()]

# Make new columns indicating what will be imputed
for col in cols_with_missing:
    data[col + '_was_missing'] = data[col].isnull()
    data[col + '_was_missing'] = data[col].isnull()

data_with_imputed_values = pd.DataFrame(my_imputer.fit_transform(data))

# Imputation removed column names; put them back
data_with_imputed_values.columns = data.columns

# Reassign index
data_with_imputed_values.index = data.index
```
