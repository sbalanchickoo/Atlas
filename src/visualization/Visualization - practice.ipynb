```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## relplot - scatter

```python
tips['tip_percent'] = tips['tip'] * 100 / tips['total_bill']
```

```python
fig = sns.relplot(x='total_bill'
                  , y='tip_percent'
                  , hue='sex'
                  , style='smoker'
                  , size='size'
                  , legend='full'
                  , kind='scatter'
                  , height=3
                  , data=tips)
# plt.figure(figsize=(15,4))
```

```python
fig = sns.relplot(x='total_bill'
                  , y='tip_percent'
                  , hue='sex'
                  , legend='full'
                  , kind='scatter'
                  , height=4
                  , row='smoker'
                  , col='size'
                  , data=tips[tips['size']<4])
```

```python
fig = sns.relplot(x='total_bill'
                  , y='tip_percent'
                  , hue='sex'
                  , style='smoker'
                  , size='size'
                  , legend='full'
                  , kind='line'
                  , data=tips)
```

```python
fig = sns.relplot(x='total_bill', y='tip_percent', kind='scatter', data=tips)
```
