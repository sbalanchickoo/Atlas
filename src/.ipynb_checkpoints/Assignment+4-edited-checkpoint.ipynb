Definitions:
* A _quarter_ is a specific three month period, Q1 is January through March, Q2 is April through June, Q3 is July through September, Q4 is October through December.
* A _recession_ is defined as starting with two consecutive quarters of GDP decline, and ending with two consecutive quarters of GDP growth.
* A _recession bottom_ is the quarter within a recession which had the lowest GDP.
* A _university town_ is a city which has a high percentage of university students compared to the total population of the city.

**Hypothesis**: University towns have their mean housing prices less effected by recessions. Run a t-test to compare the ratio of the mean price of houses in university towns the quarter before the recession starts compared to the recession bottom. (`price_ratio=quarter_before_recession/recession_bottom`)

The following data files are available for this assignment:
* From the [Zillow research data site](http://www.zillow.com/research/data/) there is housing data for the United States. In particular the datafile for [all homes at a city level](http://files.zillowstatic.com/research/public/City/City_Zhvi_AllHomes.csv), ```City_Zhvi_AllHomes.csv```, has median home sale prices at a fine grained level.
* From the Wikipedia page on college towns is a list of [university towns in the United States](https://en.wikipedia.org/wiki/List_of_college_towns#College_towns_in_the_United_States) which has been copy and pasted into the file ```university_towns.txt```.
* From Bureau of Economic Analysis, US Department of Commerce, the [GDP over time](http://www.bea.gov/national/index.htm#gdp) of the United States in current dollars (use the chained value in 2009 dollars), in quarterly intervals, in the file ```gdplev.xls```. For this assignment, only look at GDP data from the first quarter of 2000 onward.

````python
def get_list_of_university_towns():
    '''Returns a DataFrame of towns and the states they are in from the 
    university_towns.txt list. The format of the DataFrame should be:
    DataFrame( [ ["Michigan", "Ann Arbor"], ["Michigan", "Yipsilanti"] ], 
    columns=["State", "RegionName"]  )
    
    The following cleaning needs to be done:

    1. For "State", removing characters from "[" to the end.
    2. For "RegionName", when applicable, removing every character from " (" to the end.
    3. Depending on how you read the data, you may need to remove newline character '\n'. '''
    
    with open('university_towns.txt') as f:
        content = f.readlines()
    # remove whitespace characters like `\n` at the end of each line
    content = [x.strip() for x in content]
    towns = [{"RegionName": x} for x in content]
    df = pd.DataFrame(towns)

    df.loc[df["RegionName"].str.contains("\[ed"), "State"] = df["RegionName"].str.split('\[').str[0]
    df['State'] = df['State'].fillna(method='ffill')
    df.loc[(df["RegionName"].str.contains("\[")) 
           & ~(df["RegionName"].str.contains(" \(")), "RegionName"] = df["RegionName"].str.split('\[').str[0]
    df['RegionName'] = df['RegionName'].str.split(' \(').str[0].str.strip()
    df = df.loc[df["State"] != df["RegionName"], :]
    
    df.reset_index(drop=True, inplace=True)
    return df.loc[:,['State', 'RegionName']]
````

````python
def identify_recession():
    '''Returns a DataFrame idetifying recession months from year 2000 onward.
    
    The format of the returned DataFrame is :
    DataFrame( [ ["2000q1", "100", "dip", 1] ], 
    columns=["year_qtr", "gdp", "change", "recession]  )
    
    year_qtr = year and quarter as a string value
    
    gdp = gdp for that month
    
    change = change in gdp relative to previous month
    
    recession
    0 = not in recession
    1 = in recession
    '''
    
    df = pd.read_excel('gdplev.xls')
    df = df.iloc[:, 4:7]
    df.drop('Unnamed: 5', axis=1, inplace=True)
    df = df.iloc[7:, :]
    df = df.rename(columns={'Unnamed: 4': 'year_qtr', 'Unnamed: 6': 'gdp'})
    df = df.loc[df['year_qtr']>='2000', :].reset_index(drop=True).sort_values(by='year_qtr')
    df.loc[df['gdp'] < df['gdp'].shift(), 'change'] = 'dip'
    df.loc[df['gdp'] > df['gdp'].shift(), 'change'] = 'rise'
    df.loc[df['gdp'] == df['gdp'].shift(), 'change'] = 'unchanged'
    df['change'].fillna('unchanged', inplace=True)

    df['recession'] = -1

    df.loc[0,'recession'] = 0
    df.loc[1,'recession'] = 0

    df.loc[(df['recession'] == -1) &
               ((df['change'] == 'dip') & (df['change'].shift() == 'dip')), 'recession'] = 1
        
    df.loc[(df['recession'] == -1) &
           ((df['change'] == 'dip') & (df['change'].shift(-1) == 'dip')), 'recession'] = 1
    
    df.loc[(df['recession'] == -1) &
           ((df['change'].shift() == 'dip') & (df['recession'].shift() == 1)), 'recession'] = 1
    
    df.loc[(df['recession'] == -1) &
           ((df['change'].shift(2) == 'dip') & (df['recession'].shift(2) == 1)), 'recession'] = 1

    df.loc[(df['recession'] == -1), ['recession']] = 0
        
    return df
````

````python
def get_recession_start():
    '''Returns the year and quarter of the recession start time as a 
    string value in a format such as 2005q3'''
    
    df = identify_recession()
    answer = df.loc[min(df.loc[df['recession'] == 1, :].index), 'year_qtr']

    return answer
````

````python
def get_recession_end():
    '''Returns the year and quarter of the recession end time as a 
    string value in a format such as 2005q3'''
    
    df = identify_recession()

    answer = df.loc[max(df.loc[df['recession'] == 1, :].index), 'year_qtr']
    return answer
````

````python
def get_recession_bottom():
    '''Returns the year and quarter of the recession bottom time as a 
    string value in a format such as 2005q3'''
    
    df = identify_recession()

    df_rec = df.loc[df['recession'] == 1, :]
    min_gdp = df_rec['gdp'].min()
    df_bottom = df_rec.loc[df_rec['gdp'] == min_gdp, :]
    answer = df_bottom.loc[max(df_bottom.index), 'year_qtr']
    
    return answer
````

````python
def get_housing_data():
    df = pd.read_csv('City_Zhvi_AllHomes.csv')
    df.drop(['RegionID', 'Metro', 'CountyName', 'SizeRank'], axis=1, inplace=True)
    
    states = {'OH': 'Ohio', 'KY': 'Kentucky', 'AS': 'American Samoa', 'NV': 'Nevada'
              , 'WY': 'Wyoming', 'NA': 'National','AL': 'Alabama', 'MD': 'Maryland'
              , 'AK': 'Alaska', 'UT': 'Utah', 'OR': 'Oregon', 'MT': 'Montana'
              , 'IL': 'Illinois', 'TN': 'Tennessee', 'DC': 'District of Columbia', 'VT': 'Vermont'
              , 'ID': 'Idaho', 'AR': 'Arkansas', 'ME': 'Maine', 'WA': 'Washington', 'HI': 'Hawaii', 'WI': 'Wisconsin'
              , 'MI': 'Michigan', 'IN': 'Indiana', 'NJ': 'New Jersey', 'AZ': 'Arizona', 'GU': 'Guam'
              , 'MS': 'Mississippi', 'PR': 'Puerto Rico', 'NC': 'North Carolina', 'TX': 'Texas'
              , 'SD': 'South Dakota', 'MP': 'Northern Mariana Islands', 'IA': 'Iowa', 'MO': 'Missouri'
              , 'CT': 'Connecticut', 'WV': 'West Virginia', 'SC': 'South Carolina', 'LA': 'Louisiana'
              , 'KS': 'Kansas', 'NY': 'New York', 'NE': 'Nebraska', 'OK': 'Oklahoma', 'FL': 'Florida'
              , 'CA': 'California', 'CO': 'Colorado', 'PA': 'Pennsylvania', 'DE': 'Delaware', 'NM': 'New Mexico'
              , 'RI': 'Rhode Island', 'MN': 'Minnesota', 'VI': 'Virgin Islands', 'NH': 'New Hampshire'
              , 'MA': 'Massachusetts', 'GA': 'Georgia', 'ND': 'North Dakota', 'VA': 'Virginia'}
    
    df['StateFull'] = list(pd.Series(df['State']).map(states))
    df.drop('State', axis=1, inplace=True)
    df.rename(columns={'StateFull': 'State'}, inplace=True)
    
    cols = list(df.columns)
    cols_prices = []
    
    cols_metadata = ['RegionName', 'State']
    
    for col in cols:
        if (col not in cols_metadata):
            if int(col[0:4]) >= 2000:
                cols_prices.append(col)
    
    res = (df[cols_prices].groupby(pd.PeriodIndex(df[cols_prices].columns, freq='Q'), axis=1)
                  .mean()
                  .rename(columns=lambda c: str(c).lower()))
    
    return pd.concat([df[cols_metadata], res], axis=1)
````

````python
def convert_housing_data_to_quarters():
    '''Converts the housing data to quarters and returns it as mean 
    values in a dataframe. This dataframe should be a dataframe with
    columns for 2000q1 through 2016q3, and should have a multi-index
    in the shape of ["State","RegionName"].
    
    Note: Quarters are defined in the assignment description, they are
    not arbitrary three month periods.
    
    The resulting dataframe should have 67 columns, and 10,730 rows.
    '''
    df = get_housing_data()
    df = df.sort_values(by='State')
    df.set_index(['State', 'RegionName'], inplace=True)
    
    return df
````

````python
def run_ttest():
    '''First creates new data showing the decline or growth of housing prices
    between the recession start and the recession bottom. Then runs a ttest
    comparing the university town values to the non-university towns values, 
    return whether the alternative hypothesis (that the two groups are the same)
    is true or not as well as the p-value of the confidence. 
    
    Return the tuple (different, p, better) where different=True if the t-test is
    True at a p<0.01 (we reject the null hypothesis), or different=False if 
    otherwise (we cannot reject the null hypothesis). The variable p should
    be equal to the exact p value returned from scipy.stats.ttest_ind(). The
    value for better should be either "university town" or "non-university town"
    depending on which has a lower mean price ratio (which is equivilent to a
    reduced market loss).'''
    
    df_univ = get_list_of_university_towns()
    inv_states = {v: k for k, v in states.items()}
    df_univ['StateAbbrev'] = list(pd.Series(df_univ['State']).map(inv_states))
    df_univ = df_univ.loc[:, ['StateAbbrev', 'RegionName']]#.rename(columns={'StateAbbrev':'State'})
    df_univ['Uni'] = 1

    df_price = convert_housing_data_to_quarters()
    df_price.reset_index(inplace=True)
    df_price['StateAbbrev'] = list(pd.Series(df_price['State']).map(inv_states))

    df_price.reset_index(inplace=True)
    t1 = get_recession_start()
    t2 = get_recession_bottom()

    t2
    df_price = df_price.loc[:,['StateAbbrev', 'RegionName', t1, t2]]
    df_price = df_price.loc[df_price[t1].notnull() & df_price[t2].notnull(), :]
    df_price['change'] = (df_price[t2]-df_price[t1])/df_price[t1]

    df = df_price.merge(df_univ, on=['StateAbbrev', 'RegionName'], how='left')
    df['Uni'].fillna(0, inplace=True)

    df = df.loc[:, [t1, t2, 'change', 'Uni']]

    df.head()
    from scipy.stats import ttest_ind, ttest_rel

    cat1 = df.loc[df['Uni']==1, :]
    cat2 = df.loc[df['Uni']==0, :]

    uni_mean = cat1['change'].mean()
    non_uni_mean = cat2['change'].mean()

    tvalue, p = ttest_ind(cat2['change'], cat1['change'])

    if uni_mean < non_uni_mean:
        better = "non-university town"
    else:
        better = "university town"

    if p < 0.01 and (uni_mean != non_uni_mean):
        different=True
    else:
        different=False

    return different, p, better
````
