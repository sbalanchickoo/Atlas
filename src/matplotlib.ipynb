<a class="anchor" id="waffle-chart"></a>
### Waffle chart
[Back to top](#toc)

```python
tea = 224
coffee = 256
neither = 138

total = tea + coffee + neither
rows = 100
cols = 10

tea_cells = round((tea * 1000) / total)
coffee_cells = round((coffee * 1000) / total)
neither_cells = 1000 - (tea_cells + coffee_cells)

drinks = np.arange(rows*cols)
drinks[:tea_cells] = [0] * tea_cells
drinks[coffee_cells*-1:] = [2] * coffee_cells
drinks[tea_cells:coffee_cells*-1] = [1] * neither_cells

drinks_arr = np.zeros((rows, cols))

idx = 0
for r in range(rows):
    for c in range(cols):
        drinks_arr[r, c] = drinks[idx]
        idx += 1

drinks_arr = drinks_arr.T
fig = plt.figure(figsize=(100, 10))
colormap = plt.cm.coolwarm
plt.matshow(drinks_arr, cmap=colormap)
# set minor ticks
ax = plt.gca()
ax.set_xticks(np.arange(-.5, (100), 1), minor=True)
ax.set_yticks(np.arange(-.5, (10), 1), minor=True)
    
# add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)
```

```python
"""
=====
Decay
=====

This example showcases a sinusoidal decay animation.
"""


import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


def data_gen(t=0):
    cnt = 0
    
    while cnt < 1000:
        cnt += 1
        t += 0.1
        yield t, np.sin(2*np.pi*t) * np.exp(-t/10.)


def init():
    ax.set_ylim(-1.1, 1.1)
    ax.set_xlim(0, 10)
    del xdata[:]
    del ydata[:]
    line.set_data(xdata, ydata)
    return line,

fig, ax = plt.subplots()

line, = ax.plot([], [], lw=2)
ax.grid()
xdata, ydata = [], []


def run(data):
    # update the data
    t, y = data
    xdata.append(t)
    ydata.append(y)
    xmin, xmax = ax.get_xlim()
    ax.set_xlabel(data)
    if t >= xmax:
        ax.set_xlim(xmin, 2*xmax)
        ax.figure.canvas.draw()
    line.set_data(xdata, ydata)
    
    return line,

ani = animation.FuncAnimation(fig, run, data_gen, blit=False, interval=1000,
                              repeat=False, init_func=init)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Button

freqs = np.arange(2, 20, 3)

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.2)
t = np.arange(0.0, 1.0, 0.001)
s = np.sin(2*np.pi*freqs[0]*t)
l, = plt.plot(t, s, lw=2)


class Index(object):
    ind = 0

    def next(self, event):
        self.ind += 1
        i = self.ind % len(freqs)
        ydata = np.sin(2*np.pi*freqs[i]*t)
        l.set_ydata(ydata)
        plt.draw()

    def prev(self, event):
        self.ind -= 1
        i = self.ind % len(freqs)
        ydata = np.sin(2*np.pi*freqs[i]*t)
        l.set_ydata(ydata)
        plt.draw()

callback = Index()
axprev = plt.axes([0.7, 0.05, 0.1, 0.075])
axnext = plt.axes([0.81, 0.05, 0.1, 0.075])
bnext = Button(axnext, 'Next')
bnext.on_clicked(callback.next)
bprev = Button(axprev, 'Previous')
bprev.on_clicked(callback.prev)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
import matplotlib.animation as animation
from matplotlib.widgets import Slider, Button, RadioButtons
from IPython.display import HTML, Image

fig = plt.figure()
gspec = gridspec.GridSpec(17,17)

ax_norm = fig.add_subplot(gspec[2:, 0:11])
ax_gamm = fig.add_subplot(gspec[0:5, 12:])
ax_expo = fig.add_subplot(gspec[6:11, 12:])
ax_unif = fig.add_subplot(gspec[12:, 12:])
    
def init_image():
    x_norm = np.random.normal(-2.5, 1, 10000)
    x_gamm = np.random.gamma(2, 1.5, 10000)
    x_expo = np.random.exponential(2, 10000)+7
    x_unif = np.random.uniform(0,20, 10000)

    ax_norm.tick_params(direction='out', labelright=False, labelleft=True, labelbottom=True)
    ax_norm.spines['right'].set_visible(False)
    ax_norm.spines['top'].set_visible(False)
    
    ax_unif.tick_params(direction='out', labelright=False, labelleft=True, labelbottom=True)
    
    axes_side = [ax_expo, ax_gamm, ax_unif]
    for ax in axes_side:
        ax.tick_params(direction='out', labelright=False, labelleft=True, labelbottom=False)
        
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(True)
        ax.spines['top'].set_visible(False)
    generate_image(10)
    
def format_image(n: int = 1):
    ax_norm.set_title('Normal', fontsize=10)
    ax_unif.set_title('Uniform', y=0.8, fontsize=10)
    ax_gamm.set_title('Gamma', y=0.8, fontsize=10)
    ax_expo.set_title('Exponential', y=0.8, fontsize=10)
    
    ax_norm.set_xlim(0, 8)
    ax_norm.set_xticks(np.arange(0, 9, 2))
    ax_norm.set_ylim(0, 0.5)
    ax_norm.set_yticks(np.arange(0, 0.6, 0.1))
    
    axes_side = [ax_expo, ax_gamm, ax_unif]
    for ax in axes_side:
        ax.set_xlim(0, 25)
        ax.set_xticks(np.arange(0, 26, 5))
        ax.set_ylim(0, 0.5)
        ax.set_yticks(np.arange(0, 0.6, 0.1))
    ax_unif.tick_params(direction='out', labelright=False, labelleft=True, labelbottom=True)
    fig.suptitle('Various distributions by sample sizes\nn = ' + str(n), fontsize=16)
    
def update_image(n):
    if n == 100:
        anim.event_source.stop()

    ax_expo.cla()
    ax_norm.cla()
    ax_gamm.cla()
    ax_unif.cla()
    
    generate_image(n)

def generate_image(n: int = 1):
    x_expo = np.random.exponential(2, n*100)+7
    x_norm = np.random.normal(2.5, 1, n*100)
    x_gamm = np.random.gamma(2, 1.5, n*100)
    x_unif = np.random.uniform(0,20, n*100)

    ax_expo.hist(x_expo, normed=True, bins=20, alpha=0.5, color='g');
    ax_norm.hist(x_norm, normed=True, bins=20, alpha=0.5, color='c');
    ax_gamm.hist(x_gamm, normed=True, bins=20, alpha=0.5, color='r');
    ax_unif.hist(x_unif, normed=True, bins=20, alpha=0.5, color='tab:purple');

    format_image(n)
    
def init():
    ax_expo.cla()
    ax_norm.cla()
    ax_gamm.cla()
    ax_unif.cla()

def reset(event):
    samp.reset()
    init()
    anim.event_source.start()
    
def update(val):
    init()
    val = int(samp.val)
    anim.event_source.stop()
    generate_image(val)

format_image()

axcolor = 'lightgoldenrodyellow'
axsamp = plt.axes([0.25, 0.01, 0.35, 0.03], facecolor=axcolor)
samp = Slider(axsamp, 'Ad hoc sample size', 1, 100.0, valinit=1, valfmt='%0.0f')
resetax = plt.axes([0.65, 0.003, 0.3, 0.06])
button = Button(resetax, 'Continue animation', color='r', hovercolor='c')

samp.on_changed(update)
button.on_clicked(reset)

init_image()
```
