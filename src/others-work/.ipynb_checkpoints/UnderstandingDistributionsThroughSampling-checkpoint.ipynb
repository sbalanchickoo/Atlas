```python
%matplotlib notebook
import matplotlib.animation as animation
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Button

x1 = np.random.normal(0, 1, 1500)
x2 = np.random.gamma(2, 1.5, 1500)
x3 = np.random.exponential(4, 1500)
x4 = np.random.uniform(0, 10, 1500)

def update(curr):
    ax1.cla()
    ax1.axis([-4,4,0,200])
    ax1.set_title("Normal")
    ax1.hist(x1[:curr], bins=30)
    ax2.cla()
    ax2.axis([-1,15,0,200])
    ax2.set_title("Gamma")
    ax2.hist(x2[:curr], bins=40)
    ax3.cla()
    ax3.axis([-1,20,0,275])
    ax3.set_title("Exponential")
    ax3.hist(x3[:curr], bins=50)
    ax4.cla()
    ax4.axis([-1,11,0,100])
    ax4.set_title("Uniform")
    ax4.hist(x4[:curr], bins=30)
        
    
def fncstop(event):
    anim.event_source.stop()

def fncstart(event):
    anim.event_source.start()
    
    
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.subplots_adjust(right=0.875)
fig.subplots_adjust(hspace=0.35 )
bstart = Button(plt.axes([0.9, 0.8, 0.075, 0.075]), 'Start')
bstop = Button(plt.axes([0.9, 0.7, 0.075, 0.075]), 'Stop')

bstop.on_clicked(fncstop)
bstart.on_clicked(fncstart)

anim = animation.FuncAnimation(fig, update, frames=np.arange(100,1500,25), blit=True, repeat=True)
plt.show()
```
