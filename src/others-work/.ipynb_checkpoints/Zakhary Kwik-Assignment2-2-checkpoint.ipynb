```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
# Input data source
df = pd.read_csv('./data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
df['Data_Value'] = df['Data_Value']*0.1
#change datetime in the dataframe from strings to datetime class. 
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
#adds a column of date according to month and day, using a string format to produce the specific result. 
df['month-date'] = df['Date'].dt.strftime('%m-%d')
#remove leap year dates.
df = df[df['month-date'] != '02-29']

#group the values according to their max and their min values. This will leave them indexed into month-date. 
#There will only be one year's worth of values with the maximum temperature. 
maximum = df[(df['Year'] >= 2005) & df['Year'] < 2015 & (df['Element'] == 'TMAX')].groupby('month-date')['Data_Value'].max()
minimum = df[(df['Year'] >= 2005) & df['Year'] < 2015 & (df['Element'] == 'TMIN')].groupby('month-date')['Data_Value'].min()

#change the column name from Data Value to their respective variable names. 
maximum = maximum.reset_index(drop=False).rename(columns = {'Data_Value': 'Maximum'})
minimum = minimum.reset_index(drop=False).rename(columns = {'Data_Value': 'Minimum'})

#merge df with maximum and minimum so that boolean conditions can be used to find the record breaking temperatures. 
datamax = df.merge(maximum, on='month-date', how='left')
datamin = df.merge(minimum, on='month-date', how='left')

#retrieve record temperatures by boolean function
record_high = datamax[(datamax['Year'] == 2015) & (datamax['Data_Value']>datamax['Maximum'])]
record_low = datamin[(datamin['Year'] == 2015) & (datamin['Data_Value']<datamin['Minimum'])]

minimum
#this is to retrieve the years which are recorded in the dataframe.
#year = df.groupby('Year')
#year.first().index  
```

## plotting the dataframes

```python
%matplotlib notebook
import numpy as np
def list_calendar_xaxis(date1, date2): 
    x_axis = np.arange(date1, date2, dtype = 'datetime64[D]')
    return x_axis

def plot_record_temperature(axis1, axis2, colour): 
    plt.plot(axis1, axis2, linewidth = 1, color = colour)

def plot_scatter(xvalue, yvalue, colour): 
    plt.scatter(xvalue['Date'].values, yvalue['Data_Value'].values, color = colour, s = 10)
    

plt.figure(figsize = (9,5))

x = list_calendar_xaxis(date1 = '2015-01-01', date2 = '2016-01-01')
plot_record_temperature(x, maximum['Maximum'].values, 'limegreen')
plot_record_temperature(x, minimum['Minimum'].values, 'blue')

plot_scatter(record_high, record_high, colour = 'green')
plot_scatter(record_low, record_low, colour = 'blue')
#axis = list_calendar_xaxis(date1 = '2015-01-01', date2 = '2016-01-02')
#axis
```

## improving aesthetics of plot

```python
import matplotlib.dates as dates
import matplotlib.ticker as ticker

def modify_plot_area(fig): 
    #fig.xaxis.set_major_locator(dates.MonthLocator())
    fig.xaxis.set_major_formatter(dates.DateFormatter('%b'))
    for tick in fig.xaxis.get_minor_ticks(): 
        tick.tick1line.set_markersize(0) 

plt.gca().fill_between(x, maximum['Maximum'].values, minimum['Minimum'].values, facecolor='aquamarine', alpha=0.2)
plt.ylabel('Temperature (C)')
plt.title('Assignment 2: Temperature in Ann Arbor, 2005 - 2015')

# plot lines of frame, with top and right to be excluded. 
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

ax = plt.gca()
ax.legend(['Record high 2005-14', 'Record low 2005-14', 'Record high 2015', 'Record low 2015'], loc=0, frameon=False)
ax.axis(['2015/01/01','2015/12/31',-50,50])
modify_plot_area(ax)

```
