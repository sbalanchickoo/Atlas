```python
dtype = {'building_id': 'int16'
         , 'meter': 'int8'
         , 'site_id': 'int8'
         , 'primary_use': 'int8'
         , 'year_built': 'int16'
         , 'floor_count': 'int8'
         , 'air_temperature': 'float16'
         , 'cloud_coverage': 'float16'
         , 'dew_temperature': 'float16'
         , 'precip_depth_1_hr': 'float16'
         , 'sea_level_pressure': 'float16'
         , 'wind_direction': 'float16'
         , 'wind_speed': 'float16'
         , 'meter_reading': 'float16'
         , 'square_feet': 'float64'
         , 'month': 'int8'
         , 'day': 'int8'
         , 'hour': 'int8'
         , 'day_of_week': 'int8'
         , 'weekend': 'int8'
         , 'night': 'int8'
        }
df = pd.read_csv('../input/ashrae-train-test-1-0/df_label_enc.csv.gz'
                 , dtype=dtype
                 , parse_dates=['timestamp']
                 , nrows=rows)
```

```python
#plt.figure(figsize=(30,6))
cols_to_plot = ['primary_use', 'year_built', 'floor_count', 'air_temperature', 'cloud_coverage', 'dew_temperature', 'precip_depth_1_hr', 'sea_level_pressure', 'wind_direction'
                , 'wind_speed', 'square_feet']
rows = int(round(len(cols_to_plot)/2, 0))
columns = 2
fig, ax = plt.subplots(rows, columns, figsize=(18, 24))
# ax[0,0] = sns.distplot()
row = 0
column = 0
for col in cols_to_plot:
    ax[row, column].set_title(col.replace("_", " ").capitalize())
    sns.distplot(df[col], kde=True, color="b", ax=ax[row, column], axlabel="")
    if column == 1:
        row += 1
        column = 0
    else:
        column +=1

# ax[0,1].set_title('Air temperature')
# sns.distplot(df['air_temperature'], color="y", ax=ax[0, 1], axlabel="")

# # Bar chart showing average score for racing games by platform

# plt.subplot(221)
# g = sns.barplot(x=df['timestamp'], y=df['cloud_coverage'])
# #g.set_xticklabels(g.get_xticklabels(), rotation=45)


# plt.title("Cloud coverage by time")
# plt.xlabel("") # Your code here

# plt.subplot(222)
# g = sns.barplot(x=df['timestamp'], y=df['air_temperature'])
# #g.set_xticklabels(g.get_xticklabels(), rotation=45)


# plt.title("Air temperature")
# plt.xlabel("") # Your code here
plt.show
```
